```python
# Importing the pandas library with the alias pd
import pandas as pd
```

# Data Preparation

```python
# Load the dataset into a DataFrame
df = pd.read_csv('dataset.csv')
```

```python
# Display the DataFrame 'df'
df
```

```python
# Remove the 'salary' and 'salary_currency' columns from the DataFram
df.drop(columns=['salary','salary_currency'])
```

```python
# Remove the 'salary' and 'salary_currency' columns from the DataFram
df.drop(columns=['salary','salary_currency'], inplace=True)
```

```python
df
```

```python
# Check for missing values in the DataFrame 'df'
df.isnull().sum()
```

```python
# Remove rows with missing values from the DataFrame
df.dropna(inplace=True)
```

```python
# Display the DataFrame to inspect the cleaned data
df
```

```python
# Display duplicate rows in the DataFrame and keep only the first occurrence
df[df.duplicated()]
```

```python
# Iterate over each column in the DataFrame
for column in df.columns:
    # Get the unique values in the current column
    unique_values = df[column].unique()
    
    # Print the column name and its unique values
    print(f"{column}: {unique_values}")
    print()
```

```python
# Replace values in the 'experience_level' column with more descriptive labels
df['experience_level'].replace({'SE':'Senior Level/Expert',
                              'MI':'Medium Level/Intermediate',
                               'EN':'Entry Level',
                               'EX':'Executive Level'}, inplace=True)
```

```python
# Filter the DataFrame to display rows where the 'experience_level' column is 'Senior Level/Expert'
df[df['experience_level'] == 'Senior Level/Expert']
```

```python
# Filter the DataFrame to display rows where the 'experience_level' column is 'Medium Level/Intermediate'
df[df['experience_level'] == 'Medium Level/Intermediate']
```

```python
# Filter the DataFrame to display rows where the 'experience_level' column is 'Entry Level'
df[df['experience_level'] == 'Entry Level']
```

```python
# Filter the DataFrame to display rows where the 'experience_level' column is 'Executive Level'
df[df['experience_level'] == 'Executive Level']
```

# Data Analysis

```python
# Calculate the sum of values in the 'remote_ratio' column
final_sum = df['remote_ratio'].sum()
final_sum
```

```python
# Calculate the mean (average) of values in the 'remote_ratio' column
final_mean = df['remote_ratio'].mean()
final_mean
```

```python
# Calculate the standard deviation of values in the 'remote_ratio' column
standard_deviation = df['remote_ratio'].std()
standard_deviation
```

```python
# Calculate the skewness of values in the 'remote_ratio' column
skewness = df['remote_ratio'].skew()
skewness
```

```python
# Calculate the kurtosis of values in the 'remote_ratio' column
final_kurtosis = df['remote_ratio'].kurtosis()
final_kurtosis
```

```python
# Print summary statistics of sum, mean, standard deviation, skewness, and kurtosis of 'remote_ratio' column
print("Summary statistics of sum, mean, standard deviation, skewness, and kurtosis of 'remote_ratio' column")
print()
print('Sum: ', final_sum)
print('Mean: ', final_mean)
print('Standard Deviation: ', standard_deviation)
print('Skewness: ', skewness)
print('Kurtosis: ', final_kurtosis)
```

```python
# Select only columns of type 'int64'
numerical_columns = df.select_dtypes(include='int64')

# Calculate the correlation matrix for the selected numerical columns
correlation = numerical_columns.corr()

# Display the correlation matrix
correlation
```

# Data Exploration

```python
# Import the matplotlib.pyplot module with the alias plt
import matplotlib.pyplot as plt
```

```python
# Count the frequency of each job title and get the top 15 most common jobs
top_15_jobs = df['job_title'].value_counts().head(15)
top_15_jobs
```

```python
# Create a figure with a specific size
plt.figure(figsize=(10, 5))

# Create a bar plot to visualize the top 15 jobs
top_15_jobs.plot(kind='bar', color='purple')

# Set the title of the plot
plt.title('Top 15 Jobs')

# Set the label for the x-axis
plt.xlabel('Job Title')

# Set the label for the y-axis
plt.ylabel('Frequency')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Display the plot
plt.show()
```

```python
# Group the data by job title and calculate the mean salary for each group
average_salary_job = df.groupby('job_title')['salary_in_usd'].mean()

# Find the job title with the highest average salary
highest_salary_job = average_salary_job.idxmax()
highest_salary_job
```

```python
# Create a figure with a specific size
plt.figure(figsize=(20, 20))

# Create a bar plot to visualize the average salary for each job title
average_salary_job.plot(kind='bar', color='purple')

# Set the title of the plot
plt.title('Average Salary by Job Title')

# Set the label for the x-axis
plt.xlabel('Job Title')

# Set the label for the y-axis
plt.ylabel('Average Salary (USD)')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Display the plot
plt.show()
```

```python
# Calculate the mean salary for each experience level by grouping the DataFrame by 'experience_level'
salary_by_experience = df.groupby('experience_level')['salary_in_usd'].mean()
salary_by_experience
```

```python
# Create a figure with a specific size
plt.figure(figsize=(10, 5))

# Group the data by experience level and calculate the mean salary for each group,then create a bar plot to 
#visualize the average salary for each experience level
salary_by_experience.plot(kind='bar', color='purple')

# Set the title of the plot
plt.title('Average Salary by Experience Level')

# Set the label for the x-axis
plt.xlabel('Experience Level')

# Set the label for the y-axis
plt.ylabel('Average Salary (USD)')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=30)

# Display the plot
plt.show()
```

```python
# Create a figure with a specific size
plt.figure(figsize=(12, 8))

# Create the first subplot (1 row, 2 columns, position 1)
plt.subplot(1, 2, 1)

# Plot a histogram of the 'remote_ratio' column with 8 bins
plt.hist(df['salary_in_usd'], color='purple', bins=8)

# Set the title of the first subplot
plt.title('Histogram of Salary')

# Set the label for the x-axis of the first subplot
plt.xlabel('Salary')

# Set the label for the y-axis of the first subplot
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
# Create a figure with a specific size
plt.figure(figsize=(30, 1))

# Create the second subplot (1 row, 2 columns, position 2)
plt.subplot(1, 2, 2)

# Create a box plot of the 'salary_in_usd' column
plt.boxplot(df['salary_in_usd'], vert=False)

# Set the title of the second subplot
plt.title('Box plot of Salary')

# Set the label for the x-axis of the second subplot
plt.xlabel('Salary')

# Display the plot
plt.show()
```
